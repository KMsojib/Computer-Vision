```python
import cv2
from matplotlib import pyplot as plt

# Read the image
image = cv2.imread(r'D:\B - Computer Vision Library Tutorials and Projects Collection\Computer-Vision\OpenCV with Python\Resources\Photos\c1.jpg')

# Convert to different color spaces
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Display all images in one figure
images = [image_rgb, image_rgb, image_hsv, image_gray, image_lab]
titles = ['BGR Image (converted to RGB for display)', 'RGB Image', 'HSV Image', 'Gray Image', 'LAB Image']

plt.figure(figsize=(12, 8))

for i, (img, title) in enumerate(zip(images, titles)):
    plt.subplot(2, 3, i + 1)
    if len(img.shape) == 2:  # Grayscale
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img)
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
