```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

# Copying Images

```python
baboon = cv.imread("baboon.png")
if baboon is None:
    print("Photo dosen't read")
else:
    plt.figure(figsize=(5,5))
    plt.imshow(cv.cvtColor(baboon, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
A = baboon
```

```python
id(A) == id(baboon)
id(A)
```

```python
B = baboon.copy()
id(B)==id(baboon)
```

```python
baboon[:,:,] = 0
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(cv.cvtColor(baboon,cv.COLOR_BGR2RGB))
plt.title("baboon")

plt.subplot(122)
plt.imshow(cv.cvtColor(A, cv.COLOR_BGR2RGB))
plt.title("Array A")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(cv.cvtColor(baboon,cv.COLOR_BGR2RGB))
plt.title("baboon")

plt.subplot(122)
plt.imshow(cv.cvtColor(B, cv.COLOR_BGR2RGB))
plt.title("Array B")
plt.show()
```
